```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/WhiteWolf088/Data_Sets/main/pd_speech_features.csv')
df
```

```python
df.corr()['class']
```

```python
x=df.drop(['id','PPE','class'],axis=1)
x.head()
```

```python
y=df['class']
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
x_train.size,x_test.size,y_train.size,y_test.size
```

```python
x_train.shape
```

```python
import tensorflow as tf
tf.random.set_seed(7)
model_1=tf.keras.models.Sequential([
  tf.keras.layers.Dense(1000,activation='relu'),
  tf.keras.layers.Dense(500,activation='relu'),
  tf.keras.layers.Dense(200,activation='relu'),
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(50,activation='relu'),
  tf.keras.layers.Dense(10,activation='relu'),
  tf.keras.layers.Dense(1,activation='sigmoid')
])
model_1.compile(loss='binary_crossentropy',
                optimizer=tf.keras.optimizers.Adam(0.1),
                metrics=['accuracy'])

model_1.fit(x_train,y_train,
            epochs=10,
            verbose=0)
```

```python

model_1.evaluate(x_test,y_test)
```

# Improving Model accuracy

* By increasing the training Data
* train for longer
* Increasing number of hidden units

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python

from sklearn.preprocessing import MinMaxScaler
x_train_scale=MinMaxScaler().fit_transform(x_train)
x_test_scale=MinMaxScaler().fit_transform(x_test)
```

```python
tf.random.set_seed(7)
model_2=tf.keras.models.Sequential([
  tf.keras.layers.Dense(1000,activation='relu'),
  tf.keras.layers.Dense(500,activation='relu'),
  tf.keras.layers.Dense(300,activation='relu'),
  tf.keras.layers.Dense(200,activation='relu'),
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(50,activation='relu'),
  tf.keras.layers.Dense(10,activation='relu'),
  tf.keras.layers.Dense(1,activation='sigmoid')
])
model_2.compile(loss='binary_crossentropy',
                optimizer=tf.keras.optimizers.Adam(0.1),
                metrics=['accuracy'])

model_2.fit(x_train_scale,y_train,
            epochs=10,
            verbose=0)
```

```python
model_2.evaluate(x_test_scale,y_test)
```

```python
model_2.save("Parkinson's Disease Classification.h5")
```

>--> The Final accuracy of the model is 81%
